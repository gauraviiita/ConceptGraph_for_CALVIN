```python
import numpy as np
from numpy import load
import matplotlib.pyplot as plt
from PIL import Image
import glob
import cv2
```

```python
#!mkdir results10
```

```python
files = glob.glob('*.npz')
i = 0
for f in files:
    while(i<2765):
        episode = load(f)
        rgb_static = episode["rgb_static"]
        print(i)
        print(f)
        im_rgb_static = Image.fromarray(rgb_static)
       
        if(i<10):
            im_rgb_static.save("results10/frame00000"+str(i)+".jpg")
        elif (9 < i < 100):
            im_rgb_static.save("results10/frame0000"+str(i)+".jpg")
        elif (99 < i < 1000):
            im_rgb_static.save("results10/frame000"+str(i)+".jpg")
        else:
            im_rgb_static.save("results10/frame00"+str(i)+".jpg")
            
        if(i==i):
            i = i+1
            break
```
